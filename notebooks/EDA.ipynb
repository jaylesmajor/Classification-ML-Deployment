### **VISUALIZATION**

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

# Format all figures in dark mode
plt.style.use('dark_background')

# Create correlation heatmap
def plot_correlation(df):
    fig, ax = plt.subplots(figsize=(15,10))

    # Define the mask for the upper triangle to avoid duplicates
    mask = np.triu(np.ones_like(df.corr(), dtype=bool))

    # plot heatmap 
    sns.heatmap(
        df.corr().round(2),
        ax=ax,
        mask=mask,
        vmin=-1,
        vmax=1,
        annot=True,
        cmap='BrBG'
    )

    ax.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18},pad=10)
    plt.show()

plot_correlation(df)
```

```python
# Correlation heatmap as ranked
def ranked_correlation_heatmap(df, target_column):
    plt.figure(figsize=(4,8))
    heatmap= sns.heatmap(
        df.corr()[[target_column]].sort_values(by= target_column,ascending=False),
        vmin=-1,
        vmax=1,
        annot=True,
        cmap='BrBG'
    )

    heatmap.set_title('Features Correlating with Satisfied')
    plt.show()

ranked_correlation_heatmap(df, target_column='satisfied')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Create histogram and boxplot
def visualize_data(df,column_name):
    fig,axes=plt.subplots(
        figsize=(10,6),
        nrows=2,
        sharex=False
    )
    sns.set_style('dark')

    # Histogram
    sns.histplot(
        df[column_name],
        ax=axes[0],
        bins='auto',
        kde=True
    )
    axes[0].tick_params(axis='x', rotation=45)

    # Boxplot
    sns.boxplot(
        data=df,
        x=column_name,
        ax=axes[1]

    )
    plt.tight_layout()
    plt.show()

visualize_data(df, column_name='Inflight entertainment')
```
